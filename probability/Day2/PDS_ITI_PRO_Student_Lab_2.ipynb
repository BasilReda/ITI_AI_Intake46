## using `scipy`

```python
#type your code here:
x = stats.norm.rvs(size=100000)
sample_mean = np.mean(x)
sample_variance = np.var(x)
print(f"sample_mean: {sample_mean} \n sample_variance: {sample_variance}")
sns.histplot(x, kde=True)
plt.show()
```

- Compute the true "analytical" PDF using [scipy.stats.norm.pdf](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html). Plot the true PDF on the top of the histogram figure.
- Compute the CDF using [scipy.stats.norm.cdf](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), then plot it.

```python
x = stats.norm.rvs(size=100000)
x_axis = np.linspace(np.min(x), np.max(x), 1000)
true_pdf = stats.norm.pdf(x_axis)
true_cdf = stats.norm.cdf(x_axis)
plt.figure(figsize=(10, 6))
plt.hist(x, bins=50, density=True, label='Histogram', alpha=0.5)
plt.plot(x_axis, true_pdf, 'r-', linewidth=2, label='PDF')
plt.plot(x_axis, true_cdf, 'orange', linewidth=2, label='CDF')
plt.legend()
plt.title("Histogram with Analytical PDF and CDF")
plt.grid(True, alpha=0.3)

plt.show()
```

# Problem 3: Practicing the Central Limit Theorem
Use Python to simulate the Central Limit Theorem for the following random i.i.d variables and check that the higher the number of number the variables, the more the mean of subsets behave like a gaussian random variable.


```python
n_variables = 100
n_subsets = 10000
data = np.random.uniform(-1, 0, (n_subsets, n_variables))
subset_means = np.mean(data, axis=1)
plt.figure(figsize=(10, 6))
plt.hist(subset_means, bins=100, color='#1f77b4')  # Matplotlib default blue
plt.grid(True, alpha=0.5)
plt.title(f'Central Limit Theorem Simulation (N={n_variables})')
plt.xlabel('Mean of Subsets')
plt.ylabel('Frequency')
plt.show()
```
