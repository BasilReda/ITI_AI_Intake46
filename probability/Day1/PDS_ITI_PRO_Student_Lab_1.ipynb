# Problem 1

Use numpy to simulate the vector outcomes for repeating the following random experiments $1000$ times and use the relative frequency to compute the probability of the corresponding events:
    
1. **Experiment:** Flipping a fair coin, **Event:** A head occurs.
     
Observe how changing the number of repetition affects the computed probabilities.

**Hints:** 
- Use [numpy.random.randint](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html) or [numpy.random.choice](https://numpy.org/doc/stable/reference/random/generated/numpy.random.choice.html) to simulate the experiment.
- Other useful functions [numpy.arange](https://numpy.org/doc/stable/reference/generated/numpy.arange.html), [numpy.cumsum](https://numpy.org/doc/stable/reference/generated/numpy.cumsum.html), [numpy.sum](https://numpy.org/doc/stable/reference/generated/numpy.sum.html), and [matplotlib.pyplot.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
#type your code here:
n = [10 , 50 ,100 , 1000 , 5000 , 10000 , 50000 , 100000]
probabilities = []
for trial in n:
    flips = np.random.randint(0 , 2 , size = trial)
    p_e = flips.mean()
    probabilities.append(p_e)
    print(f"n: {trial} , P(e): {p_e}")

plt.plot(n , probabilities , marker = 'o')
plt.axhline(y=0.5, color='red', linestyle='--', linewidth=1)
plt.xlabel('number of flips')
plt.ylabel('probability of heads')
plt.show()
```

2. **Experiment:** Rolling a fair die, **Event:** Number 5 occurs.

```python
#type your code here:
n = [10 , 50 ,100 , 1000 , 5000 , 10000 , 50000 , 100000]
prob = []
# rolling a fair dice n times
for trial in n:
    rolls = np.random.randint(1 , 7 , size = trial)
    p_e = np.sum(rolls == 6) / trial
    prob.append(p_e)
    print(f"n: {trial} , P(6): {p_e}")

plt.plot(n , prob , marker = "o")
plt.axhline(y=1/6, color='red', linestyle='--', linewidth=1)
plt.xlabel('number of rolls')
plt.ylabel('probability of rolling a 6')
plt.show()
```

3. **Experiment:** Rolling a fair die, **Event:** A number greater than 3 occurs.

```python
#type your code here:
n = [10 , 50 ,100 , 1000 , 5000 , 10000 , 50000 , 100000]
prob = []
for trial in n:
    rolls = np.random.randint(1 , 7 , size = trial)
    p_e = np.sum(rolls >=4) / trial
    prob.append(p_e)
    print(f"n: {trial} , P(roll >= 4): {p_e}")
plt.plot(n , prob , marker = "o")
plt.axhline(y=0.5 ,  color='red', linestyle='--', linewidth=1)
plt.xlabel('number of rolls')
plt.ylabel('probability of rolling >= 4')
plt.show()
```

4. **Experiment:** Rolling two fair dice, **Event:** Sum of two dice equal 5.

```python
#type your code here:
n = [10 , 50 ,100 , 1000 , 5000 , 10000 , 50000 , 100000]
prob = []
for trial in n:
    die1 = np.random.randint(1 , 7 ,size = trial)
    die2 = np.random.randint(1 , 7 ,size = trial)
    sums = die1 + die2
    p_e = np.sum(sums == 5) / trial
    prob.append(p_e)
    print(f"n: {trial} , P(sum = 5): {p_e}")
plt.plot(n , prob , marker = "o")
plt.xlabel('number of rolls')
plt.ylabel('probability of sum = 5')
plt.show()
```

# Problem 3: Independence

Use NumPy to create simulations and compute **empirical  probability** for the following outcomes. Compare the results with the **theoretical probability**.

1. Two fair coin flips produce exactly two heads

```python
#type your code here:

# simulate 1 million tests of two fair coin flips
n = 1000000
coin1 = np.random.randint(0 , 2 , size = n)
coin2 = np.random.randint(0 , 2 , size = n)
# sums of all tests
sums = coin1 + coin2

# proportion of tests that produced exactly two heads
p_e = np.sum(sums == 2) / n
print(f"P(both coins are heads): {p_e}")
```

2. Three fair coin flips produce exactly one head

```python
#type your code here:

# simulate 1 million tests of three fair coin flips
n = 1000000
coin1 = np.random.randint(0 , 2 , size = n)
coin2 = np.random.randint(0 , 2 , size = n)
coin3 = np.random.randint(0 , 2 , size = n)

# sums of all tests
sums = coin1 + coin2 + coin3


# proportion of tests that produced exactly one head
p_e = np.sum(sums == 1) / n
print(f"P(exactly one head): {p_e}")
```

3. Three biased coin flips with P(H) = 0.6 produce exactly one head

```python
#type your code here:


# simulate 1 million tests of three biased coin flips
# hint: use np.random.choice()
n = 1000000
coin1 = np.random.choice([0 , 1] , size = n , p = [0.4 , 0.6])
coin2 = np.random.choice([0 , 1] , size = n , p = [0.4 , 0.6])
coin3 = np.random.choice([0 , 1] , size = n , p = [0.4 , 0.6])

# sums of all tests
sums = coin1 + coin2 + coin3

# proportion of tests that produced exactly one head
p_e = np.sum(sums == 1) / n
print(f"P(exactly one head): {p_e}")
```

# Problem 4:Random Variables Statistical Tools
Simulate Rolling a fair die $1000$ times. Then calculate the following:

- Sample mean
- Sample variance
- Sample standard deviation

Compare the results with the theoretical probability.

```python
#type your code here:
n = 1000
die = np.random.randint(1 , 7 , size = n)
sample_mean = die.mean()
print(f"Sample mean of die rolls: {sample_mean}")
sample_variance = die.var()
print(f"Sample variance of die rolls: {sample_variance}")
sample_std = die.std()
print(f"Sample standard deviation of die rolls: {sample_std}")
```

# Problem 5: Correlation and Covariance

The CSV file includes a hourly/daily summary for Szeged, Hungary area, between 2006 and 2016. After inspecting the data:

- Calculate the covariance and correlation coeffecient between temperature and apparent temperature
- Calculate the covariance and correlation coeffecient between temperature and humidity

```python
df = pd.read_csv('weatherHistory.csv')
temperature = df['Temperature (C)'].to_numpy().reshape(-1)
apparant_temp = df['Apparent Temperature (C)'].to_numpy().reshape(-1)
humidity = df['Humidity'].to_numpy().reshape(-1)
```

```python
plt.scatter(temperature, apparant_temp)
plt.grid()
plt.xlabel('Temperature (C)')
plt.ylabel('Apparent Temperature (C)')
```

```python
plt.scatter(humidity, temperature)
plt.grid()
plt.xlabel('Humidity')
plt.ylabel('Temperature (C)')
```

## Temperature and Apparent Temperature

```python
#type your code here:
# covariance between temperature and apparent temperature
cov = np.cov(temperature , apparant_temp)[0 , 1]
print(f"Cov: {cov}")
np_cov = np.cov(temperature , apparant_temp)
print(f"np.cov:\n{np_cov}\n")
correlation_coffiecent = np.corrcoef(temperature , apparant_temp)[0 , 1]
print(f"correlation coffiecent: {correlation_coffiecent}")
np_corr_coff = np.corrcoef(temperature , apparant_temp)
print(f"np.corrcoef:\n{np_corr_coff}") 
```

## Temperature and Humidity

```python
#type your code here:
cov = np.cov(humidity , temperature)[0 , 1]
print(f"Cov: {cov}\n")
cov = np.cov(humidity , temperature)
print(f"np(cov):{cov}\n")
corr = np.corrcoef(humidity , temperature)[0 , 1]
print(f"corr:{corr}\n")
corr = np.corrcoef(humidity , temperature)
print(f"np.corr():{corr}")
```
