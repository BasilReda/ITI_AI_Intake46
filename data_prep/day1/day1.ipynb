```python
#sepal_width
q1 = data["sepal_width"].quantile(0.25)
q3 = data["sepal_width"].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = data["sepal_width"][(data["sepal_width"] < lower_bound) | (data["sepal_width"] > upper_bound)]
print(outliers)
```

```python
data["sepal_width"] = data["sepal_width"].mask((data["sepal_width"] < lower_bound) | (data["sepal_width"] > upper_bound))
data.dropna(subset = ["sepal_width"] , inplace = True)
data.boxplot(column="sepal_width")
```
